## Ramdon Walk Generation

```python
# dt denotes how many steps per year
# xi(psi) is a ramdon brownian motion (a ramdon normal number)
import numpy as np
import pandas as pd
def gbm0(n_years=10, n_scenarios=1000, mu=0.17, sigma=0.15, steps_per_year=12,s_0=100):
    """evolution of stock price using geometric brownian motion model"""
    dt=1/steps_per_year
    n_steps=int(n_years*steps_per_year)
    xi=np.random.normal(size=(n_steps,n_scenarios))
    rets=mu*dt+sigma*np.sqrt(dt)*xi
    rets=pd.DataFrame(rets)
    # convert to prices
    prices=s_0*(1+rets).cumprod()
    return prices   
```
